```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (20, 10)  # runtime configuration parameters 20 width and 10 height
```

```python
df1 = pd.read_csv('Bengaluru House Prices.csv')
df1.head()
```

```python
df1.shape
```

```python
df1.groupby('area_type')['area_type'].agg('count')
```

```python
df2 = df1.drop(['area_type', 'society', 'balcony','availability'], axis = 'columns')
df2.head()
```

```python
# Data Cleaning
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()  # dropping all NA values
```

```python
df2['size'].unique()
```

```python
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))
# we are selecting column size, splitting it, and taking no. values to just get value instead of
# bhk or bedroom, and creating a new column out of it
```

```python
df3.head()
```

```python
df3['bhk'].unique()
```

```python
# a house having 43 bedrooms in unlikly and that at a low sq feet area
df3[df3.bhk>20]
```

```python
df3.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False;
    return True;
```

```python
df3[~df3['total_sqft'].apply(is_float)].head(10)  # gives values that can't be converted to float
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    
    if len(tokens) == 2:
        return (float(tokens[0]) + float(tokens[1])) / 2
    
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num('3090-5002')
```

```python
df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4.head(10)
```

```python
df5 = df4.copy()
```

```python
# FEATURE ENGINEERING
```

```python
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
df5.head()
```

```python
df5['location'].unique()
len(df5['location'].unique())
```

```python
# just to clean any space at beginning or end
df5.location = df5.location.apply(lambda x: x.strip())
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending = False)
location_stats
```

```python
# many have less than 10, hence we put them in other
len(location_stats[location_stats <= 10])
```

```python
location_less_than_10 = location_stats[location_stats <= 10]
location_less_than_10
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_less_than_10 else x)
len(df5.location.unique())
```

```python
# OUTLIER REMOVAL
df5.head(10)
```

```python
# 2 bhk can't have 1000 sqft. space, potential outlier
# assuming following threshold
df5[df5.total_sqft/df5.bhk < 300].head()  # per room with less than 300 sqft
```

```python
df5.shape
```

```python
df6 = df5[~(df5.total_sqft/df5.bhk < 300)]
df6.shape  # these many are now somewhat valid
```

```python
df6.price_per_sqft.describe()
```

```python
# both min and max are very unlikely
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft > (m - st)) &
                           (subdf.price_per_sqft <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out


df7 = remove_pps_outliers(df6)
df7.shape
```

```python
df7.head()
```

```python
# sometimes price of 2bhk is more than 3bhk for same area, hence we check number of these inputs

def plot_scatter(df, location):
    bhk2 = df[(df.location == location) & (df.bhk == 2)]
    bhk3 = df[(df.location == location) & (df.bhk == 3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft, bhk2.price, color = 'blue', label = '2 BHK', s = 50)
    plt.scatter(bhk3.total_sqft, bhk3.price, marker = '+', color = 'green', label = '3 BHK', s = 50)
    plt.xlabel("Total square feet area")
    plt.ylabel("Price")
    plt.title(location)
    plt.legend()


plot_scatter(df7, "Hebbal")
```

```python
# some 2bhk have higher prices than 3bhk for same area
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
df8 = remove_bhk_outliers(df7)
# df8 = df7.copy()
df8.shape
# this function removes all outliers where prices of say 1bhk is higher than 2bhk and so on
```

```python
df9 = df8[df8.bath < df8.bhk+2]
df9.shape
```

```python
# Dropping excess columns
df10 = df9.drop(['size', 'price_per_sqft'], axis = 'columns')
df10.head()
```

```python
# MODEL BUILDING
```

```python
# machine learning model can't take text data, hence we need to convert location to numeric value
# using pandas dummy method
dummies = pd.get_dummies(df10.location)
dummies.head()
```

```python
df11 = pd.concat([df10, dummies.drop('other', axis  = 'columns')], axis = 'columns')
df11.head()
```

```python
df12 = df11.drop('location', axis = 'columns')  # don't need locations now
df12.head(2)
```

```python
# removing dependent columns for model
X = df12.drop('price', axis = 'columns')
X.head()
```

```python
# dependent
y = df12.price
y.head()
```

```python
# TRAIN TEST SPLIT
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 10)
# 20% data for test rest train
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train, y_train) # train of model
lr_clf.score(X_test, y_test) # score of model
```

```python
# shufflesplit cross validation for improving score
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits = 5, test_size = 0.2, random_state = 0)
cross_val_score(LinearRegression(), X, y, cv = cv)

# performs 5 tests, 20% test rest train
```

```python
# grid search cv -> it checks all models automatically which will be best
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {}
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
def predict_price(location, sqft, bath, bhk):
    loc_index = np.where(X.columns == location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1
    return lr_clf.predict([x])[0]
```
